```python
# Splitting the data into x(the predictors) and y(the one that we want to predict)
x = df.drop(["Hours"],axis=1)
y= df["Scores"]
```

```python
# Splitting the data into train and test , 30% will be for testing  , random for shuffling 
X_train, X_test, y_train, y_test = train_test_split(x,y,test_size=0.3 ,random_state=1)
print(X_train.shape[0])
print(X_test.shape[0])
```

```python
# Prepare the model
reg = LinearRegression()
reg.fit(X_train, y_train)
```

```python
# Make the predictions
predictions= reg.predict(X_test)
print(predictions)
print(y_test.to_numpy())
```

```python
# Root mean square
res=reg.score(X_train, y_train) 
print(f"root mean square is {res}")
```

```python
# Accuracy 
r2_score = reg.score(X_test,y_test)
print(r2_score*100,'%')
```
